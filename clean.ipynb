```python
import pandas as pd
import numpy as np
```

```python
eng = pd.read_csv('data/premier-league.csv')
eng.head(30)
```

```python
#delete the euro signs in front of the values
eng['fee_cleaned'] = eng['fee_cleaned'].str.replace('€', '')
#remove the m
eng['fee_cleaned'] = eng['fee_cleaned'].str.replace('m', '')
#convert fee_cleaned to float
eng['fee_cleaned'] = eng['fee_cleaned'].astype(float)
```

```python
#dropt if the fee_cleaned is nan
eng = eng.dropna(subset=['fee_cleaned'])
eng.head(30)
```

```python
#remove if transfer_movement is out
eng = eng[eng['transfer_movement'] != 'Out']
```

```python
#group by club_name and sum the fee_cleaned
eng = eng.groupby('club_name')['fee_cleaned'].sum().reset_index()
eng = eng.sort_values(by=['fee_cleaned'], ascending=False)
eng.head(30)
```

```python
#output the csv
eng.to_csv('Prem.csv', index=False)
```

## Making the Dataframe for the Saudi League

```python
saudi = pd.read_csv('data/saudi-professional-league.csv')
saudi.head(30)
```

```python
#delete the euro signs in front of the values
saudi['fee_cleaned'] = saudi['fee_cleaned'].str.replace('€', '')
#replace the k with 000
saudi['fee_cleaned'] = saudi['fee_cleaned'].str.replace('k', '000')
#replace the m with 000000
saudi['fee_cleaned'] = saudi['fee_cleaned'].str.replace('m', '0000')
#replace the . with nothing
saudi['fee_cleaned'] = saudi['fee_cleaned'].str.replace('.', '')
#convert fee_cleaned to float
saudi['fee_cleaned'] = saudi['fee_cleaned'].astype(float)
#make it so that the fee_cleaned is in millions
saudi['fee_cleaned'] = saudi['fee_cleaned'] / 1000000
```

```python
#remove where fee_cleaned is nan
saudi = saudi.dropna(subset=['fee_cleaned'])
#remove where transfer_movement is out
saudi = saudi[saudi['transfer_movement'] != 'Out']
saudi.head(30)
```

```python
#group by club_name and sum the fee_cleaned
saudi = saudi.groupby('club_name')['fee_cleaned'].sum().reset_index()
#arrange in descending order
saudi = saudi.sort_values(by=['fee_cleaned'], ascending=False)
saudi.head(30)
```
