```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, HTML
```

```python
eng = pd.read_csv('data/premier-league.csv')
eng.head(30)
```

```python
#delete the euro signs in front of the values
eng['fee_cleaned'] = eng['fee_cleaned'].str.replace('€', '')
#remove the m
eng['fee_cleaned'] = eng['fee_cleaned'].str.replace('m', '')
#convert fee_cleaned to float
eng['fee_cleaned'] = eng['fee_cleaned'].astype(float)
```

```python
#dropt if the fee_cleaned is nan
eng = eng.dropna(subset=['fee_cleaned'])
eng.head(30)
```

```python
#remove if transfer_movement is out
eng = eng[eng['transfer_movement'] != 'Out']
```

```python
#group by club_name and sum the fee_cleaned
eng = eng.groupby('club_name')['fee_cleaned'].sum().reset_index()
eng = eng.sort_values(by=['fee_cleaned'], ascending=False)
eng.head(30)
```

```python
#output the csv
eng.to_csv('Prem.csv', index=False)
```
